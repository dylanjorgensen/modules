# Data

```python
import requests

url = 'http://geo.craigslist.org/iso/us/'
headers = {'user-agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 9_2 like Mac OS X)'}
response = requests.get(url, headers=headers)

# We are testing for the code 200
response.status_code
```

```python
# We don't need the rest of the bs4 package
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(response.content, 'html.parser')
```

```python
soup.prettify
```

# Find

### = Find All...

```python
# Find next <ul> tags
ul_list = soup.find('ul')
```

```python
# Find all <ul> tags
ul_list = soup.find_all('ul')
```

```python
# How many <ul> tags did it find?
len(ul_list)
```

### = Specify Returns...

```python
# Return single result
# ul_list[-1] # last tag
city_list_tag = ul_list[0]  # first tag

type(city_list_tag)
```

```python
# city_list = soup.find_all('ul', {'class': 'height6'}) # Only next result
city_list_set = soup.find_all('ul', {'class': 'height6'}) # Lists of specific classes

type(city_list_set)
```

# Object Types

# Loop

### = For Loop...

```python
# New dict with text as the key, and url as the value
city_dict = {}

# city['href'] is 'a' links with an 'href" <a href="#">yada</a>
for city in city_list_tag.find_all('a'):
    city_dict[city.text] = city['href']
    
city_dict
```
